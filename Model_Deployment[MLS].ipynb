#### 1. Data import and functional validation

```python
import numpy as np
import pandas as pd

DB = pd.read_csv('/content/drive/MyDrive/churn.csv')
```

```python
DB.head()
```

```python
DB.info()
```

```python
DB.nunique()
```

```python
DB['TotalCharges'].value_counts()
```

```python
DB['TotalCharges'].replace(" ", np.nan, inplace=True)
DB['TotalCharges'].value_counts()
```

```python
DB['TotalCharges'] = DB['TotalCharges'].astype('float')
DB.info()
```

```python
DB.dropna(inplace=True)
```

```python
DB.drop(['customerID'], axis=1, inplace=True)
```

```python
DB.head()
```

```python
dummies = pd.get_dummies(DB, drop_first=True)
dummies.head()
```

```python
dummies.info()
```

```python
X = dummies.drop(['Churn_Yes'], axis=1)
Y = dummies['Churn_Yes']
```

```python
from sklearn.model_selection import train_test_split
import pickle
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.7, random_state=42)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(min_samples_split=50, max_depth=12,
                               max_features=7, random_state=42, class_weight={0:0.3, 1:0.7})
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report

pred_train = model.predict(x_train)
pred_test = model.predict(x_test)

print(classification_report(y_train.ravel(), pred_train))
print(classification_report(y_test.ravel(), pred_test))
```

```python
filename = 'model.dmp'

import os
os.chdir('/content/drive/MyDrive')

pickle.dump(model, open(filename, 'wb'))
```

```python
result = saved_model.score(x_test, y_test)
print(result)
```
